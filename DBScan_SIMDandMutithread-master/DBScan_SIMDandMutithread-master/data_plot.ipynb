```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open("sample.dat", 'r')
f.readline()

a = []
for line in f:
    x, y = line.split(',')
    a.append([float(x), float(y)])
```

```python
x = [tmp[0] for tmp in a]
y = [tmp[1] for tmp in a]
```

```python
plt.scatter(x, y)
```
